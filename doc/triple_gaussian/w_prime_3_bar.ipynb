Calculate the moment analytically:

```python
w_prime_3_bar_int = sp.Integral((sp.abc.w - sym.w_bar) ** 3 * c_f.G_w, [sp.abc.w, -oo, oo])
display(sp.Eq(sym.w_prime_3_bar, w_prime_3_bar_int))
```

```python
w_prime_3_bar_int_val = w_prime_3_bar_int.doit(conds='none').simplify()
display(sp.Eq(sym.w_prime_3_bar, w_prime_3_bar_int_val))
```

The equation in the document is:

```python
display(sp.Eq(sym.w_prime_3_bar, c_f.w_prime_3_bar()))
```

```python
display(sp.factor(sp.Eq(w_prime_3_bar_int_val, c_f.w_prime_3_bar()),
                  sp.abc.alpha, sp.abc.delta))
```
