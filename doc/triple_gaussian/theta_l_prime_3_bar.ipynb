Calculate the moment analytically:

```python
theta_l_prime_3_bar_int = sp.Integral((sym.theta_l - sym.theta_l_bar) ** 3 * c_f.G_theta, [sym.theta_l, -oo, oo])
display(sp.Eq(sym.theta_l_prime_3_bar, theta_l_prime_3_bar_int))
```

```python
theta_l_prime_3_bar_int_val = theta_l_prime_3_bar_int.doit(conds='none').simplify()
display(sp.Eq(sym.theta_l_prime_3_bar, theta_l_prime_3_bar_int_val))
```

The equation in the document is:

```python
display(sp.Eq(sym.theta_l_prime_3_bar, c_f.theta_l_prime_3_bar()))
```

Setting those equations equal:

```python
display(sp.factor(sp.Eq(theta_l_prime_3_bar_int_val, c_f.theta_l_prime_3_bar()),
                  sp.abc.alpha, sp.abc.delta))
```
